```python
#Import

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams["figure.figsize"]=(12,8)
```

```python
#Read in data

df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# Seeing missing data

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Data types of columns
df.info()
```

```python
df.head()
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# To display all data
pd.set_option('display.max_rows',None)
```

```python
#drop duplicates
df.drop_duplicates()
```

```python
#Scatter plot with budget vs gross
sns.scatterplot(x=df['budget'],y=df['gross'])
plt.title('Budget Vs Gross')
plt.xlabel('Budget_Earnings')
plt.ylabel('Gross for Film')
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'magenta'})
```

```python
# Lets look at correlation
```

```python
df.corr(method='pearson') # pearson, kendall, Spearman
```

```python
correlation_matrix = df.corr(method='pearson')
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True,cmap='magma')
plt.title('Correlation Matrix for numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
# Look atcompany
df.head()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')    
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()        
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
correlation_matrix

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr().copy()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>.5]
high_corr
```
